```
#split the dataset into independent and dependent  data sets 
x = df.iloc[:,2:31].values
y = df.iloc[:,1].values
type(x)
```

```
# x contains the features that detect wheteher the patient has cancer or not
# split the data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=0)
```

```
# scale the data (features scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```
#create a function for the models
#edef models(x_train,y_train):
  #logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
print(lr)
```

```
from sklearn.ensemble import RandomForestClassifier
f= RandomForestClassifier(n_estimators=10)
f.fit(x_train,y_train)
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print(classification_report(y_test,f.predict(x_test)))
print(accuracy_score(y_test,f.predict(x_test)))
```
